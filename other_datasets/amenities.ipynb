```python
jindex=10
for ias in assesment:
    jindex+=1
    print(jindex)
#     if jindex==1:
#         dfg=pd.DataFrame(dict1)
#         print(dfg)
#         dfg.to_csv("amenities.csv")
#     else:
    res=requests.get("https://visakhapatnam.emunicipal.ap.gov.in/ptis/view/viewProperty-viewForm.action?propertyId={}&isCitizen=true#no-back-button".format(ias))
    htmlcontent = res.content
    soup = BeautifulSoup(htmlcontent, 'html.parser')
    #print(soup.prettify())
    tr=soup.find_all("tr",class_="amenities")
    if len(tr)>1:
        #     print(tr[1])
        heads=tr[1].find_all("th",class_="bluebgheadtd")
        #     print(heads)
        amenities_columns=[]
        for i in heads:
            amenities_columns.append(i.text.strip())
        #         print(amenities_columns)
        dict1={}
        print(len(amenities_columns))
        if len(amenities_columns)>=7:
            for i in amenities_columns:
                if i not in dict1:
                    dict1[i]=[]
            #             print(dict1)
            # finding values.....
            data=tr[1].find_all("td",class_="blueborderfortd")
            #         print(data)
            amenities_data=[]
            for i in data:
                amenities_data.append(i.text.strip())
            #             print(amenities_data)
            j=0
            for i in amenities_columns:
                dict1[i].append(amenities_data[j])
                j+=1
        
            dict1["assessment_number"]=[ias]
            #         dict1[assessment_number].append(dict1[assessment_number])
            print(dict1)
    
            dfg=pd.DataFrame(dict1)
            print(dfg)
            dfg.to_csv("amenities.csv",mode="a",header=False)
            print("Done")
        
    
            
print("All complete.....closing")
            
            
        
        
# 1086271319
        
```
