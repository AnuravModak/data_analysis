# PIECHART

```python
# PIE CHART FOR "Current First Half Tax"
female_first_half=sum(female["Current First Half Tax"].values)
male_first_half=sum(male["Current First Half Tax"].values)
labels=["Male","Female"]
values=[male_first_half,female_first_half]
explode = (0,0.5)
plt.pie(values,colors=["green","red"],labels=values,explode=explode,counterclock=False, shadow=True)
plt.title('Contribution to First Half Tax by gender in Rs')

plt.legend(labels,loc=3)

plt.show()

```

```python
# PIE CHART FOR "Current First Half Tax"
female_first_half=sum(female["Interest Due on Current First Half Tax Due"].values)
male_first_half=sum(male["Interest Due on Current First Half Tax Due"].values)
labels=["Male","Female"]
values=[male_first_half,female_first_half]
explode = (0,0.5)
plt.pie(values,colors=["green","red"],labels=values,explode=explode,counterclock=False, shadow=True)
plt.title('Interest amount on First Half Tax by gender in Rs')

plt.legend(labels,loc=3)

plt.show()
```

```python
# PIE CHART FOR "Current First Half Tax" on propotion
female_first_half=sum(female["Interest Due on Current First Half Tax Due"].values)/len(female["Assessment Number"].values)
male_first_half=sum(male["Interest Due on Current First Half Tax Due"].values)/len(male["Assessment Number"].values)
labels=["Male","Female"]
values=[male_first_half,female_first_half]
explode = (0,0.2)
plt.pie(values,colors=["green","red"],labels=values,explode=explode,counterclock=False, shadow=True)
plt.title('Weighted interest amount on First Half Tax by gender in Rs')

plt.legend(labels,loc=4)

plt.show()
```
