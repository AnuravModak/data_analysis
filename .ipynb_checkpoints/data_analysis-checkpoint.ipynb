# RELATIONAL ANALYSIS

```python
 def population_correlation(x,y):
    meanx=sum(x.values)/len(x)
    meany=sum(y.values)/len(y)
    num,den=0,0
    
    if len(x.values)==len(y.values):
        # numerator calculation
        a=0
        for i in range(len(x)):
            a+=x.values[i]*y.values[i]
        b=len(x)*meanx*meany
        c=meanx*(sum(y.values))
        d=meany*(sum(x.values))
        num=a+b-c-d
        # denominator calculation
        ax,ay=0,0
        for i in range(len(x)):
            ax+=x.values[i]**2
            ay+=y.values[i]**2
        bx=len(x)*((meanx)**2)
        cx=len(x)*meanx*sum(x.values)
        by=len(y)*((meany)**2)
        cy=len(y)*meany*sum(y.values)
        p1=ax+bx-cx
        p2=ay+by-cy
        den=(p1*p2)**0.5
#         print(num,den)
        return num/den
        
    else:
        raise Exception("Invalid dimension")
        
            
def linear_correlation(x,y):
    meanx=sum(x.values)/len(x)
    meany=sum(y.values)/len(y)
    num,den=0,0
    #numerator calculation
    a=0
    for i in range(len(x)):
        a+=x.values[i]*y.values[i]
#         print(a)
    num=a*len(x)-(sum(x.values)*sum(y.values))
    
    #denominator calculation
    ax,ay=0,0
    for i in range(len(x)):
            ax+=x.values[i]**2
            ay+=y.values[i]**2 
    p1=(len(x)*ax-(sum(x.values)**2))**0.5
    p2=(len(y)*ay-(sum(y.values)**2))**0.5
    den=p1*p2
    #     print(num,den)
    return num/den
    
    
    
    
```

```python
print(population_correlation(check1["Current First Half Tax"],check1["Annual Rental Value"]))
print(linear_correlation(check1["Annual Rental Value"],check1["Current First Half Tax"]))
```
